# Classification Final Project

## Main Objective of the Analysis

The objective of this analysis is to develop and evaluate predictive models that accurately classify loan applications as approved or rejected, with the goal of identifying the most effective machine learning approach for improving decision-making and reducing risk in the loan approval process.

It will assist lenders in making loan decisions based on data, reduce risk in lending, and improve efficiency by streamlining the approval outcome process through identifying key factors. Some models that are used in this analysis are very easily interpreted. Loan applicants will be able to digest the information and better understand what could make them a good applicant.

## Dataset Description

The dataset used was created for Loan Approval Prediction. For each loan applicant, the dataset outlines key features such as: 
- Loan ID
- Gender
- Marriage Status
- Number of Dependents
- Whether or not the applicant is Self-Employed
- Income of Applicant and Coapplicant
- Loan Amount and Term Requested
- Credit History
- Property Area
- Whether or not the Loan was Approved

The goal of this analysis is to discover how each of these features impact the applicants chances of loan approval. Through this process we will then be able to predict if an applicant gets approved for the loan or not.

## Data Exploration & Preparation

Through data exploration, it was discovered that many of the columns were stored as objects and therefore needed to be cast to numerical types. The nominal columns were casted arbitrarily to integers, allowing for easier use of the data. It was also evident that the Loan ID was unneeded for this analysis, so it was dropped from the dataframe. To further clean the data, rows containing NaN values were dropped. After cleaning the data, it was all scaled. 

Using the prepared data, the set was split into the training and testing sets. The same split was used thoughout this project therefore all models are evaluated under consitent conditions. 

## Models Trained

### Logistic Regression
Using Logistic Regression, we can gain a better understanding of the features. Below is a visualization of the feature coefficients:

From the graph, we can see that Credit History has the greatest positive coefficient for this dataset. This means that good credit history is the biggest indicator of whether or not an applicant will be approved. After credit history, the next largest positive coefficient is whether or not the applicant is married. If they are married, this has about a third of the positive impact that credit history has. On the other hand, loan amount and coapplicant income have the greatest negative coefficients. The larger these are, the more likely the loan will be denied. For lenders, this information can offer insight on which kind of people they should market their loans to. For example, we can extract the fact that families with kids are a prime demographic for loan approval. 

The logistic regression model achieved an accuracy of 76.8% and demonstrated strong recall for approved loans, correctly identifying approximately 94% of positive cases. However, the model struggled to correctly classify rejected applications, as reflected by low recall for the negative class. This imbalance suggests that while logistic regression provides valuable interpretability and insight into key drivers of loan approval, it may expose the lender to increased risk by misclassifying higher-risk applicants as approved. Though it is not the most accurate classifier, logistic regression provided valuable interpretability by clearly highlighting the main drivers of loan approval decisions.

### Decision Tree

One of the greatest advantages that Decision Trees offer is their interpretability. At the root node, we can see which feature is most important for classification. As we move down the tree, we can see how key features impact the loan approval decision.

As we can see, Credit History appears in the root node. Similarly to Logisitic Regression, the Decision tree has made it clear that Credit History is the most important factor in predicting if a lender will be approved for a loan. We can also see that the Applicant's Income is a driving factor in the loan approval decision.

The Decision Tree classifier achieved an accuracy of 75.8% with a notably high recall of 89.4%, indicating strong performance in identifying positive loan approval outcomes. However, this came at the cost of lower precision and overall stability, reflecting the models tendency to overfit the training data.

### Random Forest
The tuned Random Forest model achieved a training accuracy of 82.1% and a test accuracy of 76.8%, indicating strong predictive performance with a limited degree of overfitting. The observed performance gap between training and test data suggests that the model generalizes well to unseen observations while still capturing important non-linear relationships within the data.

While the Random Forest demonstrated strong performance, the difference between training and test accuracy indicates some remaining overfitting, suggesting that additional data or further regularization may improve generalization.

## Model Recommendation

Logistic regression served as a strong baseline and interpretability model, helping identify key drivers of loan approval decisions. However, due to its limited ability to capture non-linear relationships and class imbalance, it was not selected as the final model.

While the Decision Tree provided valuable interpretability and insight into non-linear decision rules, the Random Forest model delivered slightly higher accuracy and improved robustness by aggregating multiple trees. As a result, the *Random Forest* was selected as the final model due to its superior balance between predictive performance and generalization.

Compared to simpler models evaluated in this analysis, the Random Forest provided the best balance between predictive accuracy and robustness, leading to its selection as the final model.

## Key Findings & Insights

Our predictive modeling of loan approvals shows that a Random Forest model offers the best overall performance, achieving 76.8% test accuracy and effectively balancing recall and precision. Logistic regression provides strong interpretability but struggles to identify rejected applicants, while the decision tree is slightly less accurate and more prone to overfitting. Key predictors of approval include income, credit history, and debt-to-income ratio, highlighting which factors drive lending decisions. A hybrid approach using Random Forest for accurate predictions and logistic regression for transparency can guide both operational decisions and regulatory compliance, while addressing class imbalance ensures fairer assessment of all applicants.

## Limitations & Next Steps

Our analysis compared Decision Tree, Logistic Regression, and Random Forest models for predicting loan approval. Random Forest achieved the highest accuracy (~77%) and generalizes better than Decision Tree, which, while slightly less accurate, had higher recall (~89%) for approved loans. Logistic Regression performed similarly in accuracy but struggled to correctly identify rejected loans. Limitations include class imbalance, limited feature scope, potential overfitting, and reduced interpretability for complex models. Next steps involve addressing class imbalance, expanding feature selection, and optimizing hyperparameters.
